```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import face_recognition
from tqdm import tqdm
from PIL import Image, ImageEnhance

%matplotlib inline
sns.set_theme()
```

```python
df = pd.read_csv('../labels_utf8.csv')
df.head()
```

```python
results = pd.DataFrame(index=df['ID'], columns=[0.25, 0.5, 1, 2, 4, 6])

def is_face(img):
    return len(face_recognition.face_encodings(np.array(img))) > 0;

if os.path.isfile('results.csv'):
    results = pd.read_csv('results.csv')
else:
    for name in tqdm(df['ID']):
        img = Image.open(f'../../mugshots/front/front/{name}')
        for i in results.columns:
            results[i][name] = is_face(ImageEnhance.Color(img).enhance(i))

    results.to_csv('results.csv')
    
results.head()
```
